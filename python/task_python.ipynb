# Логические выражения в Python

В Python, логические выражения используются для оценки истинности условий. Они возвращают значение True или False в зависимости от результата выражения. Логические выражения могут быть созданы с помощью операторов сравнения (например, ==, !=, <, >, <=, >=) и логических операторов (например, and, or, not).

Операторы сравнения сравнивают два значения и возвращают True или False в зависимости от того, выполняется ли условие

````python
x = 5
y = 10

print(x == y)  # False, потому что x не равно y
print(x != y)  # True, потому что x не равно y
print(x < y)   # True, потому что x меньше y
print(x > y)   # False, потому что x не больше y
print(x <= y)  # True, потому что x меньше или равно y
print(x >= y)  # False, потому что x не больше или равно y
````

````python
x = 5
y = 10
z = 15

print(x < y and y < z)  # True, потому что оба условия верны
print(x > y and y < z)  # False, потому что первое условие ложно

print(x < y or y > z)   # True, потому что хотя бы одно условие верно
print(x > y or y > z)   # False, потому что оба условия ложны

print(not x == y)        # True, потому что x не равно y
print(not x != y)
````

# Строки

В Python, строки - это последовательности символов, которые используются для представления текстовой информации. Строки могут содержать буквы, цифры, знаки препинания, пробелы и т.д.

Строки в Python могут быть созданы с помощью двойных или одинарных кавычек.

````python
str1 = "Hello, World!"
str2 = 'Python is fun!'
````

Строки можно складывать

````python
str1 = "Hello, "
str2 = "World!"
str3 = str1 + str2
print(str3)  # Вывод: "Hello, World!"
````

Строки можно умножать на число

````python
str1 = "Python "
str2 = str1 * 3
print(str2)  # Вывод: "Python Python Python "
````

Можно получить срез строки с помощью индексации

````python
str1 = "Hello, World!"
print(str1[0])  # Вывод: "H"
print(str1[7])  # Вывод: "W"
````

Индексы могут быть отрицательными

````python
print(str1[-1])
print(str1[-2])
````

````python
str1 = "Hello, World!"
print(str1[0:5])  # Вывод: "Hello"
print(str1[7:12])  # Вывод: "World"
````

В Python есть множество встроенных методов для работы со строками, таких как upper(), lower(), replace(), split() и т.д.

````python
str1 = "Hello, World!"
print(str1.upper())  # Вывод: "HELLO, WORLD!"
print(str1.lower())  # Вывод: "hello, world!"
print(str1.replace("World", "Python"))  # Вывод: "Hello, Python!"
print(str1.split(","))  # Вывод: ["Hello", " World!"]
print(str1.find("World")) # Поиск подстроки в строке, возвращает индекс найденной строки
print(str1.count("o")) # посчитать количество вхождений подстроки
````

Функция len - возвращает длину строки

````python
len("abcd")
````

# List

В Python, список - это упорядоченная и изменяемая коллекция элементов, которая может содержать элементы разных типов данных. Списки могут быть созданы с помощью квадратных скобок [ ], в которых элементы разделены запятыми.

````python
list1 = [1, 2, 3, 4, 5]
list2 = ["apple", "banana", "cherry"]
list3 = [1, "apple", 3.14, True]
````

Списки могут содержать индексированные элементы, которые могут быть доступны с помощью номера индекса элемента. Индексация начинается с 0.

````python
list1 = [1, 2, 3, 4, 5]
print(list1[0])  # Вывод: 1
print(list1[2])  # Вывод: 3
````

Списки могут также содержать срезы, которые позволяют извлечь подсписок из списка. Срезы могут быть созданы с помощью индексов начала и конца и разделены двоеточием

````python
list1 = [1, 2, 3, 4, 5]
print(list1[0:3])  # Вывод: [1, 2, 3]
print(list1[2:5])  # Вывод: [3, 4, 5]
````

В Python есть множество встроенных методов для работы со списками, таких как append(), remove(), sort() и т.д.

````python
list1 = [1, 2, 3, 4, 5]
list1.append(6)
print(list1)  # Вывод: [1, 2, 3, 4, 5, 6]

list1.remove(3)
print(list1)  # Вывод: [1, 2, 4, 5, 6]

list1.sort()
print(list1)  # Вывод: [1, 2, 4, 5, 6]

list1.pop(2)
print(list1)  # Вывод: [1, 2, 5, 6]
````

Списки также могут быть объединены с помощью оператора +

````python
list1 = [1, 2, 3]
list2 = [4, 5, 6]
list3 = list1 + list2
print(list3)  # Вывод: [1, 2, 3, 4, 5, 6]
````

# Множества

В Python, множества (sets) являются неупорядоченными коллекциями уникальных элементов. Множества используются для хранения нескольких элементов в одной переменной. Множества не индексируются, поэтому к отдельным элементам множества нельзя обратиться по индексу.

Множества используются для выполнения операций над множествами, таких как объединение, пересечение, разность

Создание множества: Множество создается с помощью фигурных скобок {} или с помощью встроенной функции set()

````python
# Create a set using {}
set1 = {1, 2, 3, 4, 5}

# Create a set using set() function
set2 = set([1, 2, 3, 4, 5])
````

Добавление и удаление элементов в множестве: Элементы могут быть добавлены в множество с помощью метода add(), а могут быть удалены с помощью метода remove()

````python
# Adding an element to a set
set1.add(6)

# Removing an element from a set
set1.remove(3)
````

Операции над множествами: Некоторые операции, которые могут быть выполнены над множествами, включают:

Объединение множеств: | или union()
Пересечение множеств: & или intersection()
Разность множеств: - или difference()

````python
# Union of two sets
set3 = set1 | set2
# or
set3 = set1.union(set2)

# Intersection of two sets
set4 = set1 & set2
# or
set4 = set1.intersection(set2)

# Difference of two sets
set5 = set1 - set2
# or
set5 = set1.difference(set2)
````

# Словари

В Python, словари (dictionaries) являются упорядоченными коллекциями пар “ключ-значение”, которые используются для хранения и извлечения данных. Словари позволяют быстро и эффективно искать, добавлять и удалять элементы на основе ключей.

Создание словаря: Словарь создается с помощью фигурных скобок {} и записей “ключ:значение”.

````python
# Create a dictionary using {}
dict1 = {"name": "John", "age": 30, "city": "New York"}

# Create a dictionary using dict() function
dict2 = dict(name="John", age=30, city="New York")
````

Доступ к элементам словаря: Элементы словаря могут быть доступны по их ключам.

````python
# Access elements of a dictionary
print(dict1["name"]) # Output: John
print(dict1.get("age")) # Output: 30
````

Изменение элементов словаря: Элементы словаря могут быть изменены, удалены или добавлены новые элементы.

````python
# Change elements of a dictionary
dict1["age"] = 35

# Add new element to a dictionary
dict1["country"] = "USA"

# Remove element from a dictionary
del dict1["city"]
# or
dict1.pop("city")
````

Методы словаря: Некоторые методы, которые могут быть использованы для работы со словарями, включают:

keys(): возвращает список всех ключей в словаре.
values(): возвращает список всех значений в словаре.
items(): возвращает список всех пар “ключ-значение” в словаре.
update(): обновляет словарь новыми парами “ключ-значение”.
clear(): удаляет все элементы из словаря.

````python
# Get all keys in a dictionary
print(dict1.keys()) # Output: dict_keys(['name', 'age', 'country'])

# Get all values in a dictionary
print(dict1.values()) # Output: dict_values(['John', 35, 'USA'])

# Get all items in a dictionary
print(dict1.items()) # Output: dict_items([('name', 'John'), ('age', 35), ('country', 'USA')])

# Update a dictionary
dict1.update({"age": 40, "city": "Chicago"})

# Clear a dictionary
dict1.clear()
````

# Функции

В Python, функции являются блоками кода, которые выполняют определенную задачу. Функции позволяют повторно использовать код, делают его более читаемым и управляемым, а также позволяют передавать данные между частями программы

Создание функции: Функция создается с помощью ключевого слова def, за которым следует имя функции и параметры, заключенные в круглые скобки. Затем следует двоеточие :, а затем блок кода с отступом.

Аргументы функции - это значения, которые передаются в функцию при ее вызове.

````python
def greet(name):
    print("Hello, " + name)
````

Вызов функции: Функцию можно вызывать по ее имени и передавая ей аргументы, если это необходимо.

````python
greet("Sonya")
````

Возвращаемое значение функции: Функция может возвращать значение с помощью ключевого слова return. Если функция не возвращает значение, она возвращает None.

````python
def square(x):
    return x * x

print(square(5)) # Output: 25
````

Рекурсивные функции: Рекурсивная функция - это функция, которая вызывает саму себя. Рекурсивные функции могут быть использованы для решения многих задач, таких как вычисление факториала числа

````python
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n-1)

print(factorial(5)) # Output: 120
````

# Классы

В Python классы являются шаблонами для создания объектов. Классы определяют свойства (атрибуты) и поведение (методы) объектов, которые могут быть созданы на основе этого класса.

Создание класса: Класс создается с помощью ключевого слова class, за которым следует имя класса и двоеточие :. Затем следует блок кода с отступом, который определяет атрибуты и методы класса.

````python
class Dog:
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

    def bark(self):
        print(self.name + " says woof!")
````

В этом примере, класс Dog определяет два атрибута - name и breed, а также один метод - bark().

Создание объекта класса: Объект класса может быть создан с помощью имени класса и круглых скобок, которые содержат аргументы для метода __init__().

````python
my_dog = Dog("Fido", "Labrador")
````

В этом примере, создается объект my_dog на основе класса Dog с аргументами “Fido” и “Labrador”.

Атрибуты объекта класса: Атрибуты объекта класса определяются в методе __init__() и могут быть доступны и изменяться с помощью точечной нотации.

````python
print(my_dog.name) # Output: Fido
my_dog.breed = "Golden Retriever"
````

В этом примере, мы доступны к атрибуту name объекта my_dog и изменяем значение атрибута breed.

Методы объекта класса: Методы объекта класса определяются в определении класса и могут быть вызваны с помощью точечной нотации.

````python
my_dog.bark() # Output: Fido says woof!
````
